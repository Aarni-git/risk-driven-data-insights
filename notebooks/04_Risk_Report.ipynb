# 04 — Risk Report

This report summarizes the risk profiles discovered in Section 03.
The goal is to explain which customer segments are high-risk, why they are risky,
and how these insights can be utilized. 

## Segments included in this analysis

- income_segment (Low, Medium, High)
- percent_income_segment (Low stress, Medium stress, High stress)
- loan_amount_segment (Small, Medium, Large)

These segments were created in Section 02 and evaluated in Section 03.

## Key Risk Findings

### Income segment
- Low income customers have the highest default rate.
- High income customers have the lowest default rate.

### Percent income segment (financial stress)
- High stress customers show extremely elevated default rates.
- Medium stress customers show moderate risk.
- Low stress customers show low risk.

### Loan amount segment
- Large loans are riskier than small or medium loans.
- Medium loans have surprisingly stable risk.

### Combined segment (stress × loan size)
- High stress + Large loan is the highest-risk combination.
- Low stress + Medium loan is one of the safest combinations.


## Interpretation and Business Meaning

- High stress customers allocate a large share of income to loan payments,
  making them vulnerable to financial shocks.
- Low income customers have limited buffers, increasing default likelihood.
- Large loans amplify financial pressure and increase exposure.
- Segment combinations reveal structural risk patterns that are not visible
  when looking at single variables alone.

## Risk Profiles

### Low-risk segments
- High income + Low stress
- Medium loan + Low stress

### Medium-risk segments
- Medium income + Medium stress
- Small loan + Medium stress

### High-risk segments
- Low income + High stress
- Large loan + High stress
- High stress customers regardless of loan size

## Connection to Modeling (Section 05)

The segmentation results provide interpretable structure for the baseline ML model:
- loan_percent_income is expected to be a strong predictor.
- income and loan_amount act as stabilizing features.
- segment-level risk patterns help validate model outputs.

These insights guide feature selection and model evaluation in Section 05.


## Final Summary of Section 04

This report shows that financial stress (loan_percent_income) is the strongest driver of default risk.
Income level and loan size further shape risk profiles, especially in combination.
These findings provide a clear, interpretable foundation for credit risk modeling
and support practical decision-making in lending.
